```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#https://drive.google.com/file/d/184-p2bd3P7Fn4FjV9b7-W3tW7XRLOMKN/view?pli=1
file_id = '184-p2bd3P7Fn4FjV9b7-W3tW7XRLOMKN'
link = f'https://drive.google.com/uc?id={file_id}'
data = pd.read_csv(link)
data.head()
```

```python
sns.heatmap(data.corr())
```

```python
data = data.drop('State', axis=1) #remove the column 'State'
data = data.drop('Administration', axis=1)
data = data.drop('Marketing Spend', axis=1)
```

```python
sns.lmplot(x='R&D Spend', y='Profit', data=data)
#sns.lmplot(x='Marketing Spend', y='Profit', data=data)
```

```python
x = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
print(x)
print(y)
```

```python
from sklearn.impute import SimpleImputer

mv = [np.nan, 0]
for i in mv:
  imp_mean = SimpleImputer(missing_values=i, strategy='mean')
  imp_mean.fit(x)
  x = imp_mean.transform(x)

print(x)
```

```python
data['R&D Spend'] = x
data
```

```python
sns.lmplot(x='R&D Spend', y="Profit", data=data)
```

```python
data = data.drop(48)
```

```python
sns.lmplot(x='R&D Spend', y="Profit", data=data)
```

```python
x = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3,random_state= 8)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train[:, :] = sc.fit_transform(x_train[:, :])
x_test[:, :] = sc.fit_transform(x_test[:, :])
print(x_test)
print("----------------------------------------------------------------")
print(x_train)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train,y_train)
predictions = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

print(
  'mean_squared_error : ', mean_squared_error(y_test, predictions))
print(
  'mean_absolute_error : ', mean_absolute_error(y_test, predictions))
```

```python
print("Actual\t:\tPredictions")
for i in range(len(predictions)):
  print(f"{y_train[i]}\t:\t{predictions[i]}")
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test, predictions))
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, predictions)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Values')
plt.show()
```
